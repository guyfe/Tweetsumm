```python
import numpy as np
import pandas as pd
from catboost import CatBoostClassifier
```

```python
df = pd.read_excel('dialogs_last.xlsx')
```

```python
df
```

```python
label_name = 'has_gratitude'
```

```python
df[label_name].value_counts()
```

```python
df.loc[df[label_name] == '–ñ–∞–ª–æ–±–∞', label_name] = 1
df.loc[df[label_name] == '–í–æ–ø—Ä–æ—Å', label_name] = 0
```

```python
df = df[df[label_name].isin([0, 1])].reset_index(drop=True)
df[label_name] = df[label_name].astype('int')
```

```python
df
```

### –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
from string import punctuation

def preprocess_text(text):
    text = ''.join([el for el in text if el not in punctuation])
    text = ' '.join([ps.stem(str.lower(word)) for word in text.split()])
    return text
```

```python
df['text'] = df['text'].apply(preprocess_text)
```

### –û–±—É—á–µ–Ω–∏–µ (cross-validation)

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support
from sklearn.model_selection import RandomizedSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline

n = 5
skf = StratifiedKFold(n_splits=n, random_state=42, shuffle=True)
f1_score_results = []

for train_index, test_index in skf.split(df, df[label_name]):


        test_df = df.iloc[test_index]
        train_df, eval_df = train_test_split(df.iloc[train_index], test_size=0.25, random_state=42)

        train_df = train_df.reset_index(drop=True)
        eval_df = eval_df.reset_index(drop=True)
        test_df = test_df.reset_index(drop=True)

        clf = make_pipeline(
                TfidfVectorizer(ngram_range=(1, 2)),
                CatBoostClassifier()
            )

        clf.fit(train_df['text'], train_df[label_name])

        labels = test_df[label_name].values.astype('int')
        predictions = clf.predict(test_df['text'].values.tolist())

        precision_only, recall_only, f1_only, support_only = precision_recall_fscore_support(labels, predictions)

        f1_score_results.append(f1_only[1])
```
